### Implement Machine Learning Models

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Feature Scaling
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train, y_train)
cv_score_DT =  cross_val_score(clf, X, y, cv = 10)
cv_score_DT.mean()
```

## Random Forest Tree

```python
from sklearn.ensemble import RandomForestClassifier
rft = RandomForestClassifier(n_estimators=10, random_state = 1)
cv_scores = cross_val_score(rft, X, y, cv=10)
cv_scores.mean()
```

## SVM

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 1).fit(X_train, y_train)
cv_scores = cross_val_score(svc, X, y, cv = 10)
svc.score(X_test, y_test)
```

```python
cv_scores.mean()
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 10).fit(X_train, y_train)
cv_scores = cross_val_score(classifier, X, y, cv = 10)
```

```python
cv_scores.mean()
```

### choosing k using for loop

```python
for n in range(1, 50):
    classifier = KNeighborsClassifier(n_neighbors= n)
    cv_scores = cross_val_score(classifier, X, y, cv = 10 )
    print(n, cv_scores.mean())
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
NB = MultinomialNB()
cv_scores_NB = cross_val_score(NB, X, y, cv = 10)
cv_scores_NB.mean()
```

## Revisting SVM

```python
svc = SVC(kernel = 'rbf')
cv_scores = cross_val_score(svc, X, y, cv =10)
cv_scores.mean()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
linreg = LogisticRegression()
cv_scores_linreg = cross_val_score(linreg, X_train, y_train, cv = 10)
cv_scores_linreg.mean()
```

## Tensorflow

```python
import tensorflow as tf
from tensorflow.keras.backend import set_session
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

def create_model():
    model = Sequential()
    
    model.add(Dense(6, input_dim = 115, kernel_initializer = 'normal', activation = 'relu'))
    
    model.add(Dense(1, kernel_initializer = 'normal', activation='sigmoid'))
    
    model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
    
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
estimator = KerasClassifier(build_fn=create_model, epochs=10, verbose=0)
cv_scores = cross_val_score(estimator, X_train, y_train, cv = 10)
cv_scores.mean()
```
